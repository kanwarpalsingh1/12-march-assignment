Q1: Calculate the 95% confidence interval for a sample of data with a mean of 50 and a standard deviation of 5 using Python. Interpret the results.

```python
from scipy import stats

# Sample data
mean = 50
std_dev = 5
sample_size = 1  # Since we are dealing with a single sample, the sample size is 1

# Confidence level
confidence_level = 0.95

# Calculate the margin of error
margin_of_error = stats.norm.ppf((1 + confidence_level) / 2) * (std_dev / sample_size**0.5)

# Calculate the confidence interval
confidence_interval = (mean - margin_of_error, mean + margin_of_error)

print("95% Confidence Interval:", confidence_interval)
```

Q2: Conduct a chi-square goodness of fit test to determine if the distribution of colors of M&Ms in a bag matches the expected distribution of 20% blue, 20% orange, 20% green, 10% yellow, 10% red, and 20% brown. Use Python to perform the test with a significance level of 0.05.

```python
from scipy import stats

# Observed frequencies
observed_freq = [50, 40, 30, 15, 15, 50]  # Example data, replace with actual counts

# Expected frequencies
expected_freq = [0.20, 0.20, 0.20, 0.10, 0.10, 0.20]  # Expected proportions
total_obs = sum(observed_freq)
expected_freq = [prop * total_obs for prop in expected_freq]  # Convert proportions to counts

# Perform chi-square goodness of fit test
chi2_stat, p_val = stats.chisquare(observed_freq, expected_freq)

# Compare p-value with significance level
alpha = 0.05
if p_val < alpha:
    print("Reject null hypothesis: The observed distribution differs significantly from the expected distribution.")
else:
    print("Fail to reject null hypothesis: The observed distribution does not differ significantly from the expected distribution.")
```


Q4: A study of the prevalence of smoking in a population of 500 individuals found that 60 individuals smoked. Use Python to calculate the 95% confidence interval for the true proportion of individuals in the population who smoke.

```python
import scipy.stats as stats

# Sample size
n = 500

# Number of individuals who smoke
smokers = 60

# Calculate the sample proportion
p_hat = smokers / n

# Calculate the Z-score for 95% confidence interval
Z = stats.norm.ppf(0.975)  # For 95% confidence level (two-tailed)

# Calculate the standard error
standard_error = (p_hat * (1 - p_hat) / n) ** 0.5

# Calculate the margin of error
margin_of_error = Z * standard_error

# Calculate the confidence interval
confidence_interval = (p_hat - margin_of_error, p_hat + margin_of_error)

print("95% Confidence Interval for the proportion of smokers:", confidence_interval)
```

Q5: Calculate the 90% confidence interval for a sample of data with a mean of 75 and a standard deviation of 12 using Python. Interpret the results.

```python
import scipy.stats as stats

# Sample mean
sample_mean = 75

# Sample standard deviation
sample_std_dev = 12

# Sample size
sample_size = 1  # Since we are dealing with a single sample, the sample size is 1

# Calculate the Z-score for 90% confidence interval
Z = stats.norm.ppf(0.95)  # For 90% confidence level (two-tailed)

# Calculate the margin of error
margin_of_error = Z * (sample_std_dev / sample_size**0.5)

# Calculate the confidence interval
confidence_interval = (sample_mean - margin_of_error, sample_mean + margin_of_error)

print("90% Confidence Interval for the population mean:", confidence_interval)
```

Q6: Use Python to plot the chi-square distribution with 10 degrees of freedom. Label the axes and shade the area corresponding to a chi-square statistic of 15.

```python
import numpy as np
import matplotlib.pyplot as plt

# Chi-square distribution with 10 degrees of freedom
df = 10
x = np.linspace(0, 30, 1000)
y = stats.chi2.pdf(x, df)

# Plot the chi-square distribution
plt.plot(x, y, label='Chi-square distribution (df=10)')
plt.fill_between(x, 0, y, where=(x >= 15), color='red', alpha=0.3)  # Shade the area corresponding to chi-square statistic of 15

# Label the axes
plt.xlabel('Chi-square statistic')
plt.ylabel('Probability density')

# Show plot
plt.legend()
plt.grid(True)
plt.show()
```

Q7: A random sample of 1000 people was asked if they preferred Coke or Pepsi. Of the sample, 520 preferred Coke. Calculate a 99% confidence interval for the true proportion of people in the population who prefer Coke

```python
n = 1000

# Number of people who prefer Coke
coke_preference = 520

# Calculate the sample proportion
p_hat = coke_preference / n

# Calculate the Z-score for 99% confidence interval
Z = stats.norm.ppf(0.995)  # For 99% confidence level (two-tailed)

# Calculate the standard error
standard_error = (p_hat * (1 - p_hat) / n) ** 0.5

# Calculate the margin of error
margin_of_error = Z * standard_error

# Calculate the confidence interval
confidence_interval = (p_hat - margin_of_error, p_hat + margin_of_error)

print("99% Confidence Interval for the proportion of people who prefer Coke:", confidence_interval)
```

Q8: A researcher hypothesizes that a coin is biased towards tails. They flip the coin 100 times and observe 45 tails. Conduct a chi-square goodness of fit test to determine if the observed frequencies match the expected frequencies of a fair coin. Use a significance level of 0.05.

```python
observed_freq = [45, 55]  # Tails, Heads

# Expected frequencies for a fair coin
expected_freq = [50, 50]  # 50% tails, 50% heads

# Perform chi-square goodness of fit test
chi2_stat, p_val = stats.chisquare(observed_freq, expected_freq)

# Compare p-value with significance level
alpha = 0.05
if p_val < alpha:
    print("Reject null hypothesis: The observed frequencies differ significantly from the expected frequencies of a fair coin.")
else:
    print("Fail to reject null hypothesis: The observed frequencies do not differ significantly from the expected frequencies of a fair coin.")
```

Q9: Chi-square Test for Independence
To conduct a chi-square test for independence, we first need to calculate the expected frequencies for each cell in the contingency table under the assumption that there is no association between smoking status and lung cancer diagnosis. Then, we can calculate the chi-square statistic and compare it to the critical value at a significance level of 0.05 to determine if there's a significant association.

```python
import numpy as np
from scipy.stats import chi2_contingency

# Create the contingency table
observed = np.array([[20, 15],
                     [10, 25],
                     [15, 20]])

# Perform chi-square test for independence
chi2, p, dof, expected = chi2_contingency(observed)

print("Chi-square statistic:", chi2)
print("P-value:", p)
print("Degrees of freedom:", dof)
print("Expected frequencies:\n", expected)
```

Q10: Chi-square Test for Independence
We will conduct a chi-square test for independence to determine if there is a significant association between chocolate preference and country of origin. We will use a significance level of 0.01.

Let's perform the chi-square test using Python:

```python
import numpy as np
from scipy.stats import chi2_contingency

# Create the contingency table
chocolate_data = np.array([[120, 150, 230], [180, 200, 270]])

# Perform chi-square test for independence
chi2_stat, p_val, _, _ = chi2_contingency(chocolate_data)

# Print the results
print("Chi-square statistic:", chi2_stat)
print("p-value:", p_val)

# Check if the result is significant at 0.01 significance level
if p_val < 0.01:
    print("There is a significant association between chocolate preference and country of origin.")
else:
    print("There is no significant association between chocolate preference and country of origin.")
```

Q11: One-sample t-test
We'll conduct a one-sample t-test to determine if the sample mean of 72 is significantly different from the hypothesized population mean of 70. We'll use a significance level of 0.05.

```python
from scipy.stats import ttest_1samp

# Given data
sample_mean = 72
pop_mean = 70
sample_std = 10  # sample standard deviation (given)

# Number of observations (sample size)
n = 30

# Perform one-sample t-test
t_stat, p_val = ttest_1samp(np.random.normal(sample_mean, sample_std, n), pop_mean)

# Print the results
print("t-statistic:", t_stat)
print("p-value:", p_val)

# Check if the result is significant at 0.05 significance level
if p_val < 0.05:
    print("The population mean is significantly different from 70.")
else:
    print("There is no significant difference between the population mean and 70.")
```
